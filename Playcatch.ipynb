#<strong>1° Caso: Playcatch</strong>

Verificar se as melhorias na plataforma digital da Playcatch afetaram significativamente a navegação dos usuários, quando comparada com a plataforma concorrente.

```python
import pandas as pd
import numpy as np

df_avaliacoes = pd.DataFrame({
    'playcatch': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4],
    'concorrente': [3, 4, 3, 3, 2, 4, 3, 2, 3, 3, 4, 2, 3]
})
df_avaliacoes
```

# Hipóteses consideradas. #

$H_0: M_d = 0$

$H_1: M_d > 0 $

Onde $M_d$ é igual a diferença entre a médiana das avaliações da empresa Playcatch e da concorrente?

```python
media_playcatch = np.median(df_avaliacoes['playcatch'])
media_concorrente = np.median(df_avaliacoes['concorrente'])
dif = media_playcatch - media_concorrente

print(media_playcatch, media_concorrente, dif)
```

# Teste mais adequado:

O <strong>teste-U</strong> é mais adequado nesse contexto, pois irá comparar duas amostras <strong> independentes e não pareadas</strong> que também possuem um número pequeno de dados.

```python
from scipy.stats import mannwhitneyu
```

```python
stats, p_valor = mannwhitneyu(df_avaliacoes['playcatch'], df_avaliacoes['concorrente'], alternative='greater')
print(f'---Resultados---\nEstatística de teste: {stats:.2f}\nP-valor: {p_valor:.10f}')
```

```python
niv_sig = 0.05
if p_valor < niv_sig:
  result = 'Rejeitar a hipótese nula.'
else:
  result = 'Aceitar a hipótese nula.'

print(f'Resultado: {result}')
```

# Resultado:

O teste-U irá comparar as medianas dos dois dados. Nesse sentido, como a mediana da concorrente foi menor que a da Playcatch, podemos dizer para a empresa que, dentro de um nível de significância de 0.05, rejeitamos a hipótese nula.

# **2° Caso: Playcatch**

# Após a inclusão de um novo recurso interativo, que permite aos usuários assistir o conteúdo enquanto compartilha suas experiências em tempo real, qual foi a melhoria percebida pelos usuários?

```python
import pandas as pd

df_wt = pd.DataFrame({
    'Antes_WatchTogether': [3, 4, 3, 4, 2, 4, 3, 2, 3, 3, 4, 2, 3],
    'Depois_WatchTogether': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4]
})

df_wt
```

# Hipótese consideradas.
$H_0: M_d=0 $: A diferença entre a media das avaliações é igual a zero.

$H_1: M_d \neq 0$: A diferença entre a media das avaliações é maior do que zero.

### Como as duas amostras são dependentes e pareadas, o melhor teste é o de **Wilcoxon**.

```python
from scipy.stats import wilcoxon
```

```python
 media_antes = np.mean(df_wt['Antes_WatchTogether'])
 media_depois = np.mean(df_wt['Antes_WatchTogether'])
 print(f'Mediana antes:{media_antes:.2f}\nMediana depois:{media_depois:.2f}')
```

```python
stats, p_valor = wilcoxon(df_wt['Antes_WatchTogether'], df_wt['Depois_WatchTogether'], alternative='greater')
print(f'P-valor:{p_valor:.8f}\nEstatística de teste:{stats:.8f}')
```

```python
niv_sig = 0.05
if niv_sig > p_valor:
  resp = 'Rejeitar a hipótese nula.'
else:
  resp = 'Aceitar a hipótese nula.'

print(f'Resultado do teste: {resp}')
```
